## Fazendo o tratamento da base de dados
* Identificamos alguns tipos de problemas para tratar:
    * Tirar as strings "nao_sei"
    * Tirar elemntos como "kg" e "mol/L"
    * Mudar para inteiro os valores que na documentação para inteiro
    * Tirar valores estranhos como duração negativa e time_signature muito grande

```python
df["song_popularity"].value_counts()
```

```python
df["song_popularity"] = pd.to_numeric(df["song_popularity"],errors="coerce")
df["acousticness"] = df["acousticness"].str.replace("kg","")
df["acousticness"] = pd.to_numeric(df["acousticness"],errors="coerce")
df["danceability"] = df["danceability"].str.replace("mol/L","")
df["danceability"] = df["danceability"].apply(lambda x: float(x))
df["energy"] = pd.to_numeric(df["energy"],errors="coerce")
df["liveness"] = pd.to_numeric(df["liveness"],errors="coerce")
df["time_signature"] = pd.to_numeric(df["time_signature"],errors="coerce")
df["song_duration_ms"] = pd.to_numeric(df["song_duration_ms"],errors="coerce")
df["instrumentalness"] = pd.to_numeric(df["instrumentalness"],errors="coerce")
df["audio_mode"] = pd.to_numeric(df["audio_mode"],errors="coerce")
df["loudness"] = pd.to_numeric(df["loudness"],errors="coerce")
df["speechiness"] = pd.to_numeric(df["speechiness"],errors="coerce")
df["tempo"] = pd.to_numeric(df["tempo"],errors="coerce")
```

```python
#coloca a mediana onde tem dados faltantes, exceto na coluna "song_name"
for colunas in df.columns:
    if colunas != "song_name":
        df[colunas] = df[colunas].fillna(df[colunas].median())
```

```python
df.describe()
```

Vamos transformar as colunas que deveriam estar como inteiro para inteiro

```python
df["key"] = df["key"].apply(lambda x: int(x))
df["audio_mode"] = df["audio_mode"].apply(lambda x: int(x))
df["time_signature"] = df["time_signature"].apply(lambda x: int(x))
```

```python
df.describe().T
```

Temos uma média de duração menor que zero e um valor muito alto

```python
df = df[df["song_duration_ms"] > 0]
```

```python
df = df[df["time_signature"] < 28000000]
```

Também vendo a descrição das features no site do Spotify, vemos que alguns dados estão incoerentes como instrumentalness de valor 10 (máximo), speechiness de 97.42(também máximo) e vamos retirar esses valores e o tempo de uma música durando praticamente 30 minutos 

```python
df = df[df["speechiness"] < 2]
```

```python
df = df[df["instrumentalness"] < 2]
```

```python
df = df[df["song_duration_ms"] < 1000000]
```

A ideia é tentarmos ver também se existe uma diferença de proporcionalidade entre a popularidade da música e essas variáveis categóricas

### 1. "key"

```python
df_key = df["key"].value_counts().reset_index()
df_key.columns = ["key","Quantidade_tot"]
df_key["Prop_tot"] = df_key["Quantidade_tot"]*100/len(df)
df_key.drop(columns = "Quantidade_tot", inplace = True)
```

```python
df_80 = df[df["song_popularity"] >= 80]
df_80_key = df_80["key"].value_counts().reset_index()
df_80_key.columns = ["key","Quantidade_80"]
df_80_key["Prop_80"] = df_80_key["Quantidade_80"]*100/len(df_80)
df_80_key.drop(columns = "Quantidade_80", inplace = True)
```

```python
df_20 = df[df["song_popularity"] <= 20]
df_20_key = df_20["key"].value_counts().reset_index()
df_20_key.columns = ["key","Quantidade_20"]
df_20_key["Prop_20"] = df_20_key["Quantidade_20"]*100/len(df_20)
df_20_key.drop(columns = "Quantidade_20", inplace = True)
```

```python
df_key_m = pd.merge(df_key,df_80_key,on="key",how="outer")
df_key_m = pd.merge(df_key_m,df_20_key,on="key",how="outer")
```

```python
df_key_m.sort_values(by="key")
```

Difícil tirar muitas conclusões mas conseguimos ver uma diferença de proporções de "key" com as músicas mais populares, menos populares e o total.

### 2. "time_signature"

```python
df_times = df["time_signature"].value_counts().reset_index()
df_times.columns = ["time_signature","Quantidade_tot"]
df_times["Prop_tot"] = df_times["Quantidade_tot"]*100/len(df)
df_times.drop(columns = "Quantidade_tot", inplace = True)
```

```python
df_80_times = df_80["time_signature"].value_counts().reset_index()
df_80_times.columns = ["time_signature","Quantidade_80"]
df_80_times["Prop_80"] = df_80_times["Quantidade_80"]*100/len(df_80)
df_80_times.drop(columns = "Quantidade_80", inplace = True)
```

```python
df_20_times = df_20["time_signature"].value_counts().reset_index()
df_20_times.columns = ["time_signature","Quantidade_20"]
df_20_times["Prop_20"] = df_20_times["Quantidade_20"]*100/len(df_20)
df_20_times.drop(columns = "Quantidade_20", inplace = True)
```

```python
df_times_m = pd.merge(df_times,df_80_times,on="time_signature",how="outer")
df_times_m = pd.merge(df_times_m,df_20_times,on="time_signature",how="outer")
```

```python
df_times_m.sort_values(by="time_signature")
```

Vemos uma concecntração altíssima de "time_signature" com valor de 4, o que pode indicar uma amostragem pouco diversa. Outro ponto que podemos notar é que para as músicas populares, tanto 0 como 1 não aparecem nenhuma vez.

Para isso também vamos adicionar uma coluna "popularidade" para facilitar nos próximos plots
