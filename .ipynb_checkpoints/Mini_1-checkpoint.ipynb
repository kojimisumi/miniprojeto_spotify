## Fazendo o tratamento da base de dados
* Identificamos alguns tipos de problemas para tratar:
    * Tirar as strings "nao_sei"
    * Tirar elemntos como "kg" e "mol/L"
    * Mudar para inteiro os valores que na documentação para inteiro
    * Tirar valores estranhos como duração negativa e time_signature muito grande

```python
df["song_popularity"].value_counts()
```

```python
df["song_popularity"] = pd.to_numeric(df["song_popularity"],errors="coerce")
df["acousticness"] = df["acousticness"].str.replace("kg","")
df["acousticness"] = pd.to_numeric(df["acousticness"],errors="coerce")
df["danceability"] = df["danceability"].str.replace("mol/L","")
df["danceability"] = df["danceability"].apply(lambda x: float(x))
df["energy"] = pd.to_numeric(df["energy"],errors="coerce")
df["liveness"] = pd.to_numeric(df["liveness"],errors="coerce")
df["time_signature"] = pd.to_numeric(df["time_signature"],errors="coerce")
df["song_duration_ms"] = pd.to_numeric(df["song_duration_ms"],errors="coerce")
df["instrumentalness"] = pd.to_numeric(df["instrumentalness"],errors="coerce")
df["audio_mode"] = pd.to_numeric(df["audio_mode"],errors="coerce")
df["loudness"] = pd.to_numeric(df["loudness"],errors="coerce")
df["speechiness"] = pd.to_numeric(df["speechiness"],errors="coerce")
df["tempo"] = pd.to_numeric(df["tempo"],errors="coerce")
```

```python
for colunas in df.columns:
    if colunas != "song_name":
        df[colunas] = df[colunas].fillna(df[colunas].median())
```

```python
df.describe()
```

Vamos transformar as colunas que deveriam estar como inteiro para inteiro

```python
df["key"] = df["key"].apply(lambda x: int(x))
df["audio_mode"] = df["audio_mode"].apply(lambda x: int(x))
df["time_signature"] = df["time_signature"].apply(lambda x: int(x))
```

```python
df.describe().T
```

Temos uma média de duração menor que zero e um valor muito alto

```python
df = df[df["song_duration_ms"] > 0]
```

```python
df = df[df["time_signature"] < 28000000]
```

Também vendo a descrição das features no site do Spotify, vemos que alguns dados estão incoerentes como instrumentalness de valor 10 (máximo), speechiness de 97.42(também máximo) e vamos retirar esses valores e o tempo de uma música durando praticamente 30 minutos 

```python
df = df[df["speechiness"] < 2]
```

```python
df = df[df["instrumentalness"] < 2]
```

```python
df = df[df["song_duration_ms"] < 1000000]
```
